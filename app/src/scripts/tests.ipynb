```python
PATH = r"C:\Users\320308966\Documents\PhilipsDictionaryEYW.cs.txt"
```

```python
PHILIPS_DICTIONARY:dict[str,str] = {}
with open(PATH, 'r') as file:
   # Read one line at a time
   for line in file:
    line = line.strip()
    if "new DictionaryTag(Instance" in line:
        parts = line.split(',')
        if len(parts) >= 2:
            tag = parts[1].strip()[2:]
            vr = parts[2].strip()[3:]
            name = parts[3].strip()[1:-3]
            PHILIPS_DICTIONARY[f"({tag})"] = {
                "tag": f"({tag})",
                "vr": vr,
                "name": name,
                "vm": "1",
                "version": "PHILIPS"
            }
   
```
